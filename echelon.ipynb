***Echelon forms***

```python
import numpy as np
import sympy as sp
```

***Solve the following system of equations:<br>
x+4y+3z = 11<br>
2x+10y+7z = 27<br>
x+y+2z = 5.<br>***


```python
from sympy import Matrix

M=Matrix([
    [1,4,3,11],
    [2,10,7,27],
    [1,1,2,5]
])

print(f'This is our matrix {M}')
print(f'This is our row-echelon form{M.echelon_form()}')

#Extract the rref component only
m_rref,m_pivots=M.rref()

print(f'This is our reduced row-echelon form{m_rref}')
print(f'Soultion vector:{m_rref.col(-1)}')
```

***Plot in 3D to see the solution visually***

```python
import matplotlib.pyplot as plt
import numpy as np

# Create the figure and 3D axes
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Define x and y ranges
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Define the three plane equations (z = ...)
Z1 = (11-X + 4*Y) / 3
Z2 = (27-2*X-10*Y)/7
Z3 = (5-X-Y) / 2

# Plot each surface with different colors and transparency (alpha)
ax.plot_surface(X, Y, Z1, alpha=0.5, color='royalblue', label='x+4y+3z=11')
ax.plot_surface(X, Y, Z2, alpha=0.5, color='orange', label='2x+10y+7z = 27')
ax.plot_surface(X, Y, Z3, alpha=0.5, color='green', label='x+y+2z = 5')

# MARK THE POINT
px, py, pz = -2, 1, 3
ax.scatter(px, py, pz, color='red', s=100, zorder=5)
ax.text(px, py, pz, "  Intersection (-2, 1, 3)", size=10, zorder=10, fontweight='bold')

# Label axes
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.title('Visualization of 3D Planes')
plt.show()
```
