**Exercise 1.3.2 Use elementary operations to find the point (x, y) that lies on both lines<br> x + 3y = 1<br>
4x−y = 3.**

```python
''' Import necessary libraries '''
import numpy as np
```

```python
# Enter the coefficients matrix
A = np.array([[1,3],[4,-1]])
print (A)
```

```python
# Enter the constants vector
B = np.array([1,3])
print (B)
```

```python
# Solve the equations
k = np.linalg.solve(A,B)
print(f'Solution{k}')
```

***Plot the equations in matplotlib to confirm<br>***
x + 3y = 1<br>
4x−y = 3<br>

```python
import matplotlib.pyplot as plt

x=np.linspace(-1,1,100)
y1=(1-x)/3
y2=(4*x-3)

plt.plot(x,y1,label='X+3Y=1')
plt.plot(x,y2,label='4X-Y=3')
plt.title('Solving Linear Equations')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
# Label the point automatically
plt.scatter(k[0], k[1], color='red', zorder=5)
plt.annotate(f'({k[0]:.2f}, {k[1]:.2f})',
                 xy=(k[0], k[1]), xytext=(5, 5),
                 textcoords='offset points', fontweight='bold')
plt.legend()
plt.show()
```

**Exercise 1.3.3 Use elementary operations to determine whether the three lines <br>x + 2y = 1 <br> 2x − y = 1 <br> 4x+3y = 3 <br>have a common point of intersection. If so, find the point, and if not, tell why they don’t have such a common point of intersection.**

```python
# Enter the Coefficient matrix
A=np.array([[1,2],[2,-1],[4,3]])
print(A)
```

```python
# Enter the constants vector
B=np.array([1,1,3])
print(B)
```

***The following will not work since linalg works only with square matrices***

```python
# Solve the system
try:
    k = np.linalg.solve(A, B)
    print("Solution found:", k)
except np.linalg.LinAlgError as err:
    print(f"Linear algebra error caught: {err}")
```

***Exercise 1.3.4 Do the three planes, <br>x+y−3z = 2 <br> 2x+y+z = 1 <br> 3x+2y−2z = 0 <br> have a common point of intersection? If so, find one and if not, tell why there is no such point.***


```python
# Coefficient matrix
A=np.array([[1,2,3],[1,1,2],[-3,1,-2]])
print(A)
# Results vector
B=np.array([2,1,0])
print(B)

try:
    k = np.linalg.solve(A, B)
    print("Solution found:", k)
except np.linalg.LinAlgError as err:
    print(f"Linear algebra error caught: {err}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create the figure and 3D axes
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Define x and y ranges
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Define the three plane equations (z = ...)
Z1 = (X + Y - 2) / 3
Z2 = 1 - 2*X - Y
Z3 = (3*X + 2*Y) / 2

# Plot each surface with different colors and transparency (alpha)
ax.plot_surface(X, Y, Z1, alpha=0.5, color='royalblue', label='x+y-3z=2')
ax.plot_surface(X, Y, Z2, alpha=0.5, color='orange', label='2x+y+z=1')
ax.plot_surface(X, Y, Z3, alpha=0.5, color='green', label='3x+2y-2z=0')

# Label axes
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.title('Visualization of 3D Planes')
plt.show()
```

***Exercise 1.3.5 Solve the following system of equations by back substitution.<br>
x + 3y − 2z = 5<br>
y + 3z = 4<br>
z = 1.***

```python
A=np.array([[1,3,-2],[0,1,3],[0,0,1]])
print(A)
B=np.array([5,4,1])
print(B)

try:
    k = np.linalg.solve(A, B)
    print("Solution found:", k)
except np.linalg.LinAlgError as err:
    print(f"Linear algebra error caught: {err}")
```

***Example 1.22: Solving a system of equations: an infinite set of solutions<br>Solve the following system of equations:<br>
3x−y+5z = 8<br>
y−10z = 1<br>
6x−y = 17<br>***

```python
A=np.array([[3,-1,5],[0,1,-10],[6,-1,0]])
print(A)
B=np.array([8,1,17])
print(B)

try:
    k = np.linalg.solve(A, B)
    print("Solution found:", k)
except np.linalg.LinAlgError as err:
    print(f"Linear algebra error caught: {err}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create the figure and 3D axes
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Define x and y ranges
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Define the three plane equations (z = ...)
Z1 = (8-3*X+Y)/8
Z2 = (1-Y)/-10
Z3 = (17-6*X)/6

# Plot each surface with different colors and transparency (alpha)
ax.plot_surface(X, Y, Z1, alpha=0.5, color='royalblue', label='3x−y+5z = 8')
ax.plot_surface(X, Y, Z2, alpha=0.5, color='orange', label='y−10z = 1')
ax.plot_surface(X, Y, Z3, alpha=0.5, color='green', label='6x−y = 17')

# Label axes
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.title('Visualization of 3D Planes')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create the grid
x = np.linspace(0, 6, 20)
y = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x, y)

# Equations
Z1 = (8 - 3*X + Y) / 5
Z2 = (Y - 1) / 10
# Plane 3 is vertical (6x - y = 17). We'll plot it as a vertical surface.
z_range = np.linspace(-2, 2, 20)
X_v, Z_v = np.meshgrid(x, z_range)
Y_v = 6*X_v - 17

# Plot surfaces
ax.plot_surface(X, Y, Z1, alpha=0.4, color='blue', label='Eq 1')
ax.plot_surface(X, Y, Z2, alpha=0.4, color='orange', label='Eq 2')
ax.plot_surface(X_v, Y_v, Z_v, alpha=0.4, color='green', label='Eq 3')

ax.set_xlabel('X'); ax.set_ylabel('Y'); ax.set_zlabel('Z')
plt.title('Visualization of Dependent 3D Planes')
plt.show()
```
